```python
#import libraries
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# loading dataset
df = pd.read_csv('Crime_Data_preprocessed.csv')
df
```

```python
#Convert all categorical to numerical
df[df.select_dtypes(['object']).columns]=df.select_dtypes(['object']).apply(lambda x:pd.factorize(x)[0])
print(df)
```

```python
#Seperate the X and Y columns
import numpy as np
X=df.values
Y=df['Vict Age']
```

```python
#Drop target column from features
X=np.delete(X,4,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_scaler=scaler.fit_transform(X)
X_scaler[:1]
```

```python
#convert high dimensions to low dimensions
from sklearn.decomposition import PCA
pca=PCA(n_components=7)  #I changes many times n_components values but at 48 it gives best resuults
X_pca=pca.fit_transform(X_scaler)
print(X_pca.shape)
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X_pca,Y,test_size=0.20 ,random_state=20)
```

### Training

```python
from keras.models import Sequential
from keras.layers import Dense, Input

# Initialize the model
model = Sequential()

# Add input layer and first hidden layer using the Input layer
model.add(Input(shape=(X_Train.shape[1],)))
model.add(Dense(units=64, activation='relu'))

# Add additional hidden layers
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=64, activation='relu'))

# Add the output layer
model.add(Dense(units=1, activation='linear'))  # Linear activation for regression

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])

```

```python
model.fit(X_Train, Y_Train, epochs=50, validation_split=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assume you have trained your model and split your data into X_train and Y_train

# Predict the training data
Y_train_pred = model.predict(X_Train)

# Calculate metrics
mae = mean_absolute_error(Y_Train, Y_train_pred)
mse = mean_squared_error(Y_Train, Y_train_pred)
r2 = r2_score(Y_Train, Y_train_pred)
print(f"Mean Absolute Error on training set: {mae:.2f}")
print(f"Mean Squared Error on training set: {mse:.2f}")
print("Accuracy")
print(f"R² Score on training set: {r2:.2f}")
```

### Testing

```python
Y_pred = model.predict(X_Test)
```

```python
loss, mae = model.evaluate(X_Test, Y_Test)
print(f"Mean Absolute Error on test data: {mae}")

print("Accuracy")
r2 = r2_score(Y_Test, Y_pred) 
print(f"R² Score on test data: {r2}")
```
